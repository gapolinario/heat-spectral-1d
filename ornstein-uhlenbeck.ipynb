$$
\dot u = -cu + \sigma \eta(t)
$$
with c > 0

$$
u(t_{n+1}) = e^{-ch} u(t_n) + \frac{\sigma}{\sqrt{2c}} \sqrt{1-e^{-2ch}} \chi_n
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 100.
T = 1.
c=1./T
sigma=1.
TotalTime = 1000.
numsteps = int(TotalTime/h)
size = 10000
taxis = np.arange(numsteps)*h
```

```python
def var_analyt(t):
    return sigma**2*.5*(1.-np.exp(-2.*c*t))/c
```

```python
u = np.empty((numsteps,size))
u[0,:] = np.zeros(size)
for i in range(1,numsteps):
    u[i,:] = u[i-1,:]*np.exp(-c*h) + sigma/np.sqrt(2.*c)*np.sqrt(1.-np.exp(-2.*c*h))*np.random.normal(0,1,size=size)
```

```python
# mean x time, avg. over ensemble
plt.plot(np.mean(u,axis=1)) # ens. average
plt.axhline(np.mean(u),color='red') # ens. and time average
plt.axhline(0.,color='k',linestyle='dashed') # analytical
plt.show()
```

```python
# var x time, avg. over ensemble
plt.plot(taxis,np.var(u,axis=1))
plt.plot(taxis,var_analyt(taxis),color='k',linestyle='dashed')
plt.show()
```
